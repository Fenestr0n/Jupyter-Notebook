````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file = open("dataset_address.csv", "r", encoding='utf-8')
data = file.read()

street_types = ["улица", 'переулок', 'шоссе', 'площадь', 'бульвар', 'проезд', 'тупик', 'просек', 'проспект', 'набережная', 'аллея', 'квартал', 'линия']
word_count = {}

for stype in street_types:
    word_count[stype] = data.count(stype)
file.close()

ds = pd.Series(word_count)
ds.sort_values(inplace=True)

fig = plt.figure(figsize=(12, 8))
ax = plt.axes()
ax = sns.barplot(x = ds.values, y = ds.index)
plt.xlabel("Количество вхождений")
plt.ylabel("Набор слов")

for i in ax.containers:
    ax.bar_label(i)
````
