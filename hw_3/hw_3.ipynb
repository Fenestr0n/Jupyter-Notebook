```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

link = "http://www.radioscanner.ru/base/index.php?action=list&sortBy=1&page="
frequencies = []

for num_page in range(0, 884):
    url = link + str(num_page)
    page = requests.get(url).text
    soup = BeautifulSoup(page, "lxml")
    table = soup.find_all("tr", attrs={"class" : ["tbCel1", "tbCel2"]})
 
    for row in table:
      columns = row.findAll("td", attrs={"class" : "caption1"})

      frequencies.append(
            {
             "frequency" : columns[0].find("a").get_text(strip=True),
             "place" : columns[1].contents[0].strip(),
             "observation_date" : columns[1].find("small").get_text(strip=True),
             "modulation" : columns[2].find("strong").get_text(strip=True),
             "type_signal" : columns[2].contents[2].strip(),
             "radio_service" : columns[3].find("small").get_text(strip=True),
             "affiliation" : columns[4].get_text(strip=True),
             "call_sign" : columns[5].find("small").get_text(strip=True),
            })

df = pd.DataFrame(frequencies)
df.to_csv("freq.csv", index=False)
df
```
