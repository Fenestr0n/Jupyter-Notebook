## Дана функция:

### f(x) = -5*x**2 + 10*x - 150

- Определить корни
- Найти интервалы, на которых функция возрастает
- Найти интервалы, на которых функция убывает
- Построить график
- Вычислить вершину
- Определить промежутки, на котором f > 0
- Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')

fx = -5*x**2 + 10*x - 150
fx
```

```python
plot(fx);
```

```python
solve(fx, x)
```

```python
x1, x2 = solve(fx)
round(x1.evalf(), 3), round(x2.evalf(), 3)
```

```python
solve(fx > 0)
```

```python
solve(fx < 0)
```

```python
solve(diff(fx) < 0)
```

```python
solve(diff(fx) > 0)
```
